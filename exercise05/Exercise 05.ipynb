```python
import points
```

```python
from points import distance
```

```python
# read airport data
data = points.files.read_csv('airports.csv')
```

```python
# get latitude and longitude
lat = points.util.to_float(points.util.get_attribute(data,'latitude'))
```

```python
lon = points.util.to_float(points.util.get_attribute(data,'longitude'))
```

```python
# make those points
points = list(zip(lat, lon))
```

```python
# Q1
# nearest neighbor calculation and output them as list
nndlist = []
for i, point in enumerate(points):
    nndlist.append(distance.nearest_airport(i,points))
nndlist = list(nndlist)
nndlist
```

```python
# Q2
# sort the output list by distance and pick the biggest distance which means the most isolated
def getKey(item):
    return item[2]
nndlist.sort(key = getKey, reverse = True)
isoairport = nndlist[0]
isoairport
```

```python
# Q3 
# compare those distance between two points and output the biggest distance
dis = 0
for i, point in enumerate(points):
        for j, point in enumerate(points):
            d_ip = distance.euclidean_distance(points[i], point)
            if d_ip != 0:
                if d_ip > dis:
                    dis = d_ip
                    fn_id = j
                    origin_id = i
furthest = (origin_id,fn_id,dis)
furthest
```

```python
# Q4
# use haversine method to calculate the distance
nndlist_h = []
for i, point in enumerate(points):
    nndlist_h.append(distance.nearest_airport_h(i,points))
nndlist_h = list(nndlist_h)
nndlist_h
```

```python
def getKey(item):
    return item[2]
nndlist_h.sort(key = getKey, reverse = True)
isoairport = nndlist_h[0]
isoairport
```

```python
dis = 0
for i, point in enumerate(points):
        for j, point in enumerate(points):
            d_ip = distance.haversine(points[i], point)
            if d_ip != 0:
                if d_ip > dis:
                    dis = d_ip
                    fn_id = j
                    origin_id = i
print(origin_id,fn_id,dis)
furthest = (origin_id,fn_id,dis)
furthest
```

```python
# Q5
# compare with two list
list(set(nndlist) - set(nndlist_h))
```
